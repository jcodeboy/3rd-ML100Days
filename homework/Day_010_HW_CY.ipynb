```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path='./data/'
df_train=pd.read_csv(data_path + 'house_train.csv.gz')
df_train.head()
```

```python
train_Y=np.log1p(df_train['SalePrice'])
df=df_train.drop(['Id','SalePrice'],axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features=[]
for dtype,feature in zip(df.dtypes,df.columns):
    if dtype=='int64' or dtype=='float64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features :{num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df=df[num_features]
df=df.fillna(-1)
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
print(train_Y.shape)
print(df.shape)
```

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x=df['1stFlrSF'],y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
MMScaler=MinMaxScaler()
temp_X=MMScaler.fit_transform(df)
estimator=LinearRegression()
cross_val_score(estimator,temp_X,train_Y,cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
q1=np.quantile(df['1stFlrSF'],0.25)
q3=np.quantile(df['1stFlrSF'],0.75)
iqr=q3-q1
b_h=q3+1.5*iqr
b_l=q1-1.5*iqr

df['1stFlrSF']=df['1stFlrSF'].clip(b_l,b_h)
sns.regplot(x=df['1stFlrSF'],y=train_Y)

temp_X=MMScaler.fit_transform(df)
cross_val_score(estimator,temp_X,train_Y,cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
keep_indexs=((df['1stFlrSF']>b_l) & (df['1stFlrSF']<b_h))
df15=df[keep_indexs]
sns.regplot(x=df15['1stFlrSF'],y=train_Y[keep_indexs])
# 做線性迴歸, 觀察分數
temp_X=MMScaler.fit_transform(df[keep_indexs])
estimator=LinearRegression()
cross_val_score(estimator,temp_X,train_Y[keep_indexs],cv=5).mean()
```
